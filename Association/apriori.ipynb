# Apriory Algorithm

```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = pd.read_csv("Market_Basket_Optimisation.csv", sep=",", header=None)

dataset.head()
```

```python
# dataset.info()
dataset.isnull().sum().sum()
```

```python
dataset.shape
```

```python
transaction = []
for i in range(0, dataset.shape[0]):
    temp = []
    for j in range(0, dataset.shape[1]):
        x =[str(dataset.values[i,j])]
        if x != ["nan"]:
            temp.append(str(dataset.values[i,j]))
    temp = np.array(temp)
    transaction.append(temp)
transaction = np.array(transaction)
```

```python
transaction.shape
```

```python
transaction
```

```python
df = pd.DataFrame(transaction, columns=["items"])
df["count"] = 1
```

```python
df.head()
```

```python
df.shape
```

```python
te = TransactionEncoder()
te_ary = te.fit_transform(transaction)
data = pd.DataFrame(te_ary, columns=te.columns_)

data
```

# Apriori

## Support

```python
fi = apriori(data, min_support=0.05, use_colnames=True)
fi.sort_values(by = "support", ascending=False)
```

## Confidence

```python
association_rules(fi, metric="confidence", min_threshold=0.05)
```

## Lift

```python
association_rules(fi, metric="lift", min_threshold=1.2)
```
